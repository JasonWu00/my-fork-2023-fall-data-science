# Rough Guide a v1 model.
0. Make a model to predict PTS 
0. For v1, dont use `2P, 3P, FG, or FT`.  Instead use `2PA, 3PA, FTA`.  
0. Include any other cols use see fit.  


````python
indp_vars = ["2PA", "3PA", "FTA"]
dpdt_vars = "PTS"

x_train, x_test, y_train, y_test = train_test_split(df[indp_vars], df[dpdt_vars], test_size=0.2)

pts_model = LinearRegression()

pts_model.fit(x_train, y_train)

y_pred = pts_model.predict(x_test) 

rmse_error = rmse(y_test, y_pred)
r_squared = metrics.r2_score(y_test, y_pred)
print("RMSE error: ", rmse_error)
print("R-squared error: ", r_squared)
````

````python
# What if we accounted for all of them?

indp_vars = ["2PA", "3PA", "FTA", "AST", "oppPTS"]
dpdt_vars = "PTS"

x_train, x_test, y_train, y_test = train_test_split(df[indp_vars], df[dpdt_vars], test_size=0.2)

pts_model = LinearRegression()

pts_model.fit(x_train, y_train)

y_pred = pts_model.predict(x_test) 

rmse_error = rmse(y_test, y_pred)
r_squared = metrics.r2_score(y_test, y_pred)
print("RMSE error: ", rmse_error)
print("R-squared error: ", r_squared)
````

````python
# Copied from lecture material.

df["pts_diff"] = df["PTS"] - df["oppPTS"]

X = df["pts_diff"]
y = df["W"]

X = sm.add_constant(X)

print(X)
print(y)

model_wins = sm.OLS(y, X).fit()

print(model_wins.params)

coef_y_int = model_wins.params['const']
coef_runs_diff = model_wins.params['pts_diff']

print(coef_y_int, coef_runs_diff)

WINS = 40

min_runs_diff = (WINS - coef_y_int) / coef_runs_diff
min_runs_diff
````
